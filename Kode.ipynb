```python
import scipy.stats as stats

#Buat fungsi untuk menghitung korelasi dan menguji korelasi 
def lwise_corr_pvalues(df30, to_csv= False, file_name = None, method= None):
    df30 = df30.dropna(how= 'any')._get_numeric_data() 
    dfcols = pd.DataFrame(columns=df30.columns) 
    rvalues = dfcols.transpose().join(dfcols, how='outer') 
    pvalues = dfcols.transpose().join(dfcols, how='outer') 
    length = str(len(df30))
   
    if method == None:
        test = stats.pearsonr 
        test_name = "Pearson"
    elif method == "spearman": 
        test = stats.spearmanr 
        test_name = "Spearman Rank" 
    elif method == "kendall": 
        test = stats.kendalltau 
        test_name = "Kendall's Tau-b" 

    for r in df30.columns: 
        for c in df30.columns: 
            rvalues[r][c] = round(test(df30[r], df30[c])[0], 4) 

    for r in df30.columns: 
        for c in df30.columns: 
            pvalues[r][c] = format(test(df30[r], df30[c])[1], '.4f') 

    if to_csv == False: 
        print("Correlation test conducted using list-wise deletion",  
            "\n", 
            "Total observations used: ",length, "\n", "\n",  
            f"{test_name} Correlation Values", "\n", rvalues, 
            "\n", 
            "Significant Levels", "\n", pvalues) 

    if to_csv == True: 
        print("Correlation test conducted using list-wise deletion", 
            "\n", 
            "Total observations used: ",length, "\n", "\n",  
            f"{test_name} Correlation Values", "\n", rvalues, 
            "\n", 
            "Significant Levels", "\n", pvalues) 

        file = open(file_name, 'a') 
        file.write("Correlation test conducted using list-wise deletion" + "\n" + f"{test_name} correlation values" + "\n")
        file.write("Total observations used: " + length + "\n") 
        file.close() 
        rvalues.to_csv(file_name, header= True, mode= 'a') 
        file = open(file_name, 'a') 
        file.write("p-values" + "\n") 
        file.close() 
        pvalues.to_csv(file_name, header= True, mode= 'a') 
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Standardisasi data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df)

# Terapkan PCA
pca = PCA(n_components=2)  # Mengurangi ke 2 dimensi
data_pca = pca.fit_transform(data_scaled)

# Tentukan jumlah cluster menggunakan Elbow Method
wcss = []  # Within-cluster sum of squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(data_pca)
    wcss.append(kmeans.inertia_)

# Visualisasi Elbow Method
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Jumlah Cluster')
plt.ylabel('WCSS')
plt.show()

# Setelah menentukan jumlah cluster (misalnya 3), terapkan K-Means
n_clusters = 3  # Ganti sesuai dengan hasil dari Elbow Method
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
clusters = kmeans.fit_predict(data_pca)

# Tambahkan hasil kluster ke DataFrame
df['Cluster'] = clusters

# Mengurutkan cluster
df_sorted = df.sort_values(by='Cluster').reset_index(drop=True)
df_sorted

# Visualisasi hasil klustering
plt.figure(figsize=(10, 5))
scatter = plt.scatter(data_pca[:, 0], data_pca[:, 1], c=df['Cluster'], cmap='rainbow', alpha=0.7)

# Menambahkan legend untuk setiap cluster dengan warna yang sesuai
cluster_labels = np.unique(df['Cluster'])  # Mengambil label cluster unik
handles = [plt.Line2D([0], [0], marker='o', color='w', 
                       markerfacecolor=scatter.cmap(scatter.norm(i)), markersize=10) for i in cluster_labels]

plt.legend(handles, [f'Cluster {i}' for i in cluster_labels], title='Cluster')

plt.title('Hasil Clustering K-Means dengan PCA')
plt.xlabel('Komponen PCA 1')
plt.ylabel('Komponen PCA 2')
plt.show()

# Menghitung rata-rata Total Skor untuk setiap cluster
cluster_mean_scores = df_sorted.groupby('Cluster')['Total Skor'].mean()

print("Rata-rata Total Skor setiap cluster:")
print(cluster_mean_scores)
```
